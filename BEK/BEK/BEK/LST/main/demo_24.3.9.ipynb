```julia
using LinearAlgebra
using ProgressMeter
using DelimitedFiles
using NonlinearEigenproblems
```

```julia
ω=0.008
N=59
Ro=1
Co=2
α_all=nothing
β_all=nothing
R_all=nothing
α_all=[α_all;1+1*im]
β_all=[β_all;0.001]
R_all=[R_all;251]
R=488
@showprogress 1 "Conputing..." for R=251:1:500
    for  β=0.001:0.001:0.15 
    A0,A1,A2,A3,A4=KEB_SpatialMode.KEB_LST_ALL("Vonkarmen copy 2.txt",N,ω,β,R,Ro,Co)
    nep = PEP([A0,A1,A2,A3,A4]); #Create a PEP object
    sc=10
    nep1 = shift_and_scale(nep,scale=sc);
    mult_scale = norm(nep1.A[end]);
    nep2 = PEP(nep1.A ./ mult_scale);
    norm.(get_Av(nep2))
    λ1,v1 = iar(nep2,σ=0.03,neigs=10,maxit=500);
    λ_orig = sc*λ1
    C=filter(x->-0.1<imag(x)<0 && 0<real(x)<0.7,λ_orig)
    if C==[]
        α=1 + 1 * im
    else
        D=findmax(imag,C)
        α=C[D[2],1]
    end
     if α_all[end] == 1+1*im
        if imag(α) <-0.01 ||  real(α)>0.7 || real(α) < 0
        α=1 + 1 *im
        end
    end
    α_all=[α_all;α]
    β_all=[β_all;β]
    R_all=[R_all;R]
    end
end
α_all=α_all[3:end]
β_all=β_all[3:end]
R_all=R_all[3:end]
data_imag=[R_all β_all imag(α_all)]
data_real=[R_all β_all real(α_all)]
data=[R_all β_all α_all]
writedlm("Vonkarmen_spa_0.008_imag.dat",data_imag ,'\t')
writedlm("Vonkarmen_spa_0.008_real.dat",data_real ,'\t')
writedlm("Vonkarmen_spa_0.008.dat",data,'\t')
```
