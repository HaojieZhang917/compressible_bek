```julia
include("LST_BEK_vonKarmen.jl")
using LinearAlgebra
using Plots
using NonlinearEigenproblems
using ProgressMeter
using DelimitedFiles
```

```julia
ω=0
N=39
Ro=0
Co=2
α_all=nothing
β_all=nothing
R_all=nothing
α_all=[α_all;-1-1*im]
β_all=[β_all;0.001]
R_all=[R_all;251]
@showprogress 1 "Conputing..." for R=90:2:300
    for  β=0.041:0.001:0.25
    A0,A1,A2,A3,A4=KEB_SpatialMode.KEB_LST_OS_SPA("Ekman.txt",N,ω,β,R)
    nep = PEP([A0,A1,A2,A3,A4]); #Create a PEP object
    sc=10
    nep1 = shift_and_scale(nep,scale=sc);
    mult_scale = norm(nep1.A[end]);
    nep2 = PEP(nep1.A ./ mult_scale);
    λ1,v1 = iar(nep2,σ=0.05,neigs=10,maxit=500);
    λ_orig = sc*λ1
    C=filter(x->0<imag(x)<0.25 && 0.05<real(x)<1,λ_orig)
    if C==[]
        α=-1-1*im
    else
        D=findmin(imag,C)
        α=C[D[2],1]
    end
     if α_all[end]==-1-1*im
        if imag(α)>0.05
        α=-1-1*im
        end
    end
    if α_all[end]!=-1-1*im
        if abs(imag(α)-imag(α_all[end]))>0.05
            α=-1-1*im
        end
    end
    α_all=[α_all;α]
    β_all=[β_all;β]
    R_all=[R_all;R]
end
end
α_all=α_all[3:end,1]
β_all=β_all[3:end,1]
R_all=R_all[3:end,1]
data_imag=[R_all β_all imag(α_all)]
data_real=[R_all β_all real(α_all)]
data=[R_all β_all α_all]
writedlm("Ekman_spa_os_om=0_im.dat",data_imag ,'\t')
writedlm("Ekman_spa_os_om=0_re.dat",data_real ,'\t')
writedlm("Ekman_spa_os_om=0.dat",data,'\t')
```
