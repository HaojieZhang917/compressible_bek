```julia
using LinearAlgebra
using DelimitedFiles
```

```julia
f = open("Ekman_R=100_imag.dat", "r" )
n = countlines( f )
seekstart( f )
data = zeros(n,6)
for i = 1:n
    al,be,om = split( readline( f ), "   " ) 
    data[i,1] = parse(Float64,al)
    data[i,2] = parse(Float64,be)
    data[i,3] = parse(Float64,om)
end
```
