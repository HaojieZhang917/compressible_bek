```julia
 using BoundaryValueDiffEq
 using Plots
 using DifferentialEquations
```

```julia
xspan = (0,10)
initial_vec = [0,0,0]
function plate_blasius(du, u , p , x)
    f = u[1]
    g = u[2]
    h = u[3]
    du[1] = g
    du[2] = h
    du[3] = -0.5 * f * h
end
```

```julia
function bc1!(residual, u , p ,t )
    residual[1] = u[end][2] - 1
    residual[2] = u[1][2] - 0 
    residual[3] = u[1][1] - 0
end
```

```julia
prob = BVProblem(plate_blasius , bc1! , initial_vec , xspan)
sol1 = solve(prob, MIRK4(),dt = 0.01)
plot(sol1)
```
