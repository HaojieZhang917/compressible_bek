```julia
include("CRD_STA.jl")
using Plots
using LinearAlgebra
using NonlinearEigenproblems
using ProgressMeter
using DelimitedFiles
using PyCall
```

```julia
for Tw = 0.8 
N_cheb = 149
Mr = 0.6
gamma = 1.4
sigma = 0.72
be_start = -0.05
be_end = 0.22
be_step = 0.001
R_start = 26
R_end = 100
R_step = 0.1
Ro = 1
Co = 0
global u0,v0,w0,f,q,D,D2,x = baseflow_var(N_cheb,Ro,Co)
H,T = T_ca(Mr,f,q,w0,gamma,Tw)
F,G,H,T,rho,z = interp(u0,v0,H,T,x,N_cheb,"sim")
lam = - (2/3) * T
kappa = (1/sigma) * T
for omega = -0.1 
    global initial_i_1 = []
    global initial_r_1 = []
    global initial_i_2 = []
    global initial_r_2 = []
    global tempvec_i = [0]
    GR = 0
    writedlm("output_eig.dat",initial_r_1)
    writedlm("output_GR_$omega _$Tw.dat",initial_r_1)
    writedlm("output_GR_data_$omega _$Tw.dat",initial_r_1)
    for be = be_start : 0.5*be_step : be_end
        Ma = Mr/R_start
        A0,A1,A2 = Spatial_mode_BEK(F,G,H,rho,lam,kappa,T,sigma,gamma,R_start,Ma,omega,be,N_cheb,Ro,Co)
        nep = PEP([A0,A1,A2]); 
        eigval,eigvec = iar(nep,σ = 0.1 , neigs = 1 ,maxit = 500,tol=1e-10)
        eigval = conj(eigval)
        global tempvec_i = [tempvec_i;imag(eigval[1])]
        global tempvec_r = [tempvec_i;real(eigval[1])]

        open("output_eig.dat", "a") do io
            write(io,"be=$be,eig=$eigval\n")
        end
        if imag(eigval[1]) * tempvec_i[end - 1,1] < 0 && abs.(imag(eigval[1]))<0.03
            global initial_i_1 = [omega R_start be 0]
            global initial_r_1 = [omega R_start be 0.6]
            global initial_i_2 = [omega R_start be imag(eigval[1])]
            global initial_r_2 = [omega R_start be real(eigval[1])]
            break
        end
    end
        global total_r_1 = initial_r_1
        global total_i_1 = initial_i_1
        global total_r_2 = initial_r_2
        global total_i_2 = initial_i_2
        global beta = initial_i_2[end,3]
    for R = R_start  : R_step : R_end
        Ma = Mr/R
        A0,A1,A2 = Spatial_mode_BEK(F,G,H,rho,lam,kappa,T,sigma,gamma,R,Ma,omega,beta,N_cheb,1,0)
        nep = PEP([A0,A1,A2]); 
        # eigval_1,eigvec = iar(nep , σ = total_r_1[end,4] - im * total_i_1[end,4] , neigs = 1 ,maxit = 500,tol=1e-10)
        # eigval_1 = conj(eigval_1)
        eigval_2,eigvec = iar(nep , σ = total_r_2[end,4] - im * total_i_2[end,4] , neigs = 1 ,maxit = 500,tol=1e-10)
        eigval_2 = conj(eigval_2)
        GR_temp = imag(eigval_2[1]) * R_step
        GR = GR + abs(GR_temp)
        open("output_GR_$omega _$Tw.dat", "a") do io
        write(io,"R = $R , eigmode2 = $eigval_2 , GR = $GR\n")
        end
        open("output_GR_data_$omega _$Tw.dat", "a") do io
        write(io,"$beta\t$R\t$GR\n")
        end
        # if imag(eigval_1[1])<0
        #     total_i_1 = [omega R be imag(eigval_1[1])]
        #     total_r_1 = [omega R be real(eigval_1[1])]
        # else 
        #     total_i_1 = [omega R be 0]
        #     total_r_1 = [omega R be 0.6]
        # end
        if imag(eigval_2[1]) < 0
            total_i_2 = [omega R beta imag(eigval_2[1])]
            total_r_2 = [omega R beta real(eigval_2[1])]
        end
    end
end
end
```
