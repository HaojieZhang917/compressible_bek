```julia
using DelimitedFiles
```

```julia
omega = -0.1
Tw = 0.8
original = read("output_GR_$omega _$Tw.dat",String)
header = "Zone T = \"Tw=$Tw, omega=$omega\""
```

```julia
for Tw = 0.8 : 0.1 : 1.2
    for omega = -0.1 : -0.001 : -0.11
        original = read("output_GR_data_$omega _$Tw.dat",String)
        header = "Zone T = \"Tw=$Tw, omega=$omega\""
        open("output_GR_data_$omega _$Tw.dat", "w") do io
            write(io,header * "\n" * original)
        end
    end
end
```
